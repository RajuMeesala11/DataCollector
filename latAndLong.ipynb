```python
import pandas as pd
from geopy.geocoders import Nominatim
from geopy.exc import GeocoderTimedOut

# Initialize geocoder
geolocator = Nominatim(user_agent="geoapiExercises")

def get_lat_lon(address):
    try:
        location = geolocator.geocode(address)
        if location:
            return location.latitude, location.longitude
        else:
            return None, None
    except GeocoderTimedOut:
        return None, None

```

```python
# Assuming 'Address' column contains the addresses
df['Latitude'], df['Longitude'] = zip(*df['Address'].apply(get_lat_lon))

# Save the updated DataFrame to a new CSV
df.to_csv("addresses_with_lat_lon.csv", index=False)

print("Latitude and Longitude added successfully.")
```

```python
# Read CSV file
csv_file = "datacenters.csv"  # Replace with your file name
df = pd.read_csv(csv_file)

df.head()
```

```python
df_address = df[['Address']].copy()
```

```python
# Single address geocoding
for address in df_address['Address']:
    try:
        location = geolocator.geocode(address, timeout=10)
        if location:
            df_address.loc[df_address['Address'] == address, 'Latitude'] = location.latitude
            df_address.loc[df_address['Address'] == address, 'Longitude'] = location.longitude
        else:
            df_address.loc[df_address['Address'] == address, 'Latitude'] = None
            df_address.loc[df_address['Address'] == address, 'Longitude'] = None
    except GeocoderTimedOut:
        df_address.loc[df_address['Address'] == address, 'Latitude'] = None
        df_address.loc[df_address['Address'] == address, 'Longitude'] = None
 
```
